### [1] 'stock.csv'와 'my_stock.csv' 파일 확인하기
#### 코스닥 주식관련 데이터셋 'stock.csv' 파일(12행*8열)을 읽어 온 데이터프레임 df이다
- 종목코드는 식별코드로 6자리로 이루어져 있고 문자를 포함하고 있어 문자열타입이다   
- 종목크드는 끝자리가 보통주 일경우 ‘0’, 우선주인 경우는 ‘5’, 'K', 'M'등으로 구성되어 있다
- 참고 사이트 : http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201   

```python
df = pd.read_csv('stock.csv', index_col='종목코드')
df
```

### [2] df에서 다음 조건에 따른 결과를 r2에 대입하라
- 업종별 '종가', '시가', '거래대금' 각 컬럼에 대하여 최고값과 최저값의 편차로 집계한다
- (예) 유통업 '종가'의 경우 유통업 종가의 최고값 - 최저값을 구한다
- 다음의 r2의 결과를 참고하라


```python
# [2] 코드 작성
r2 = df.groupby('업종명')[['종가', '시가', '거래대금']].apply(lambda x: x.max()-x.min())
# r2 = df.groupby('업종명')[['종가', '시가', '거래대금']].agg(lambda x: x.max()-x.min())
# r2 = df.pivot_table(index='업종명', aggfunc=lambda x: x.max()-x.min()).reindex(['종가', '시가', '거래대금'], axis = 1)
r2
```

### [3] df에서 다음 조건에 따른 결과를 r3에 대입하라
- 업종별로 '거래대금'은 max와 min집계, '시가'는 min집계, '종가'는 max집계 한다
- 다음의 r3의 결과를 참고하라


```python
# [3] 코드 작성
r3 = df.groupby('업종명').agg({'거래대금':['max','min'], '시가':'min', '종가':'max'})
# r3 = df.pivot_table(index='업종명', aggfunc={'종가':'max', '시가':'min', '거래대금':['max','min']})
r3
```

### [4] df에서 다음 조건에 따른 결과를 처리하라
- '거래량'컬럼에서 같은 업종별 최고 거래량값으로 채운 Series를 '최고거래량'컬럼으로 추가하라
- 다음 결과를 참고하라


```python
# [4] 코드 작성
df['최고거래량'] = df.groupby('업종명')['거래량'].transform('max')
df
```

### [5] df에서 다음 조건에 따라 처리하여 r5에 대입하라
- 업종별로 등락률의 최저값이 +인 업종 그룹의 행들을 추출하여 r5에 대입한다
- x는 업종별 등락률의 최저값을 집계한 결과이다. r5의 구현시 x를 참고하라
- 다음 r5의 결과를 참고하라


```python
x = df.groupby('업종명')['등락률'].min() # 업종별 등락률의 최저값(확인용)
x

# [5] 코드 작성
r5 = df.groupby('업종명').filter(lambda x: x['등락률'].min()>0)
r5
```
