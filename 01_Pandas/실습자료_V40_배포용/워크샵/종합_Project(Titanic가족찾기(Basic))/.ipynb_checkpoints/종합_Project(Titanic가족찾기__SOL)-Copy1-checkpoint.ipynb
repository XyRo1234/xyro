## [문제 2] 같은 가족 조건에 부합하는 데이터 추출하기 1
### 2-1. df에서 Last와 Ticket별 인원 집계를 다음 조건에 따라 처리하여 df2에 대입하라
1. 'Last'와 'Ticket'별 인원집계한다. 가장 많은 인원순으로 배치된다.
- 성(Last)과 티켓이 같다면 가족일 가능성이 크다라는 조건에 부합한 자료를 추출하기 위함이다
2. 'Last'와 'Ticket' 인덱스를 컬럼으로 재배치한다. 
3. 인원 집계한 컬럼명은 'cnt'로 부여한다
4. 단 처리하는 과정에서 동일한 cnt일 경우 Last와 Ticket의 순서는 다를수 있다. 
- 다음은 df2의 결과 일부이다. 
  

### 2-2. df2['cnt'] 컬럼에서 10개만 추출한 Series를 r2에 대입한다
- 다음은 r2의 인쇄 결과이며 검증용 코드에서 r2을 인쇄하여 검증한다


```python
# [문제 2]
#--[아래에 코드 작성]--------------------------------------------------

df2 = df[['Last', 'Ticket']].value_counts().reset_index().rename(columns={0:'cnt'})
# df2 = df[['Last', 'Ticket']].value_counts().to_frame('cnt').reset_index()
# df2 = df.groupby(['Last','Ticket']).size().sort_values(ascending=False).reset_index().rename(columns={0:'cnt'})
r2 = df2['cnt'].head(10)

#--[검증용코드, 수정금지]--------------------------------------------------
df2.head()
print(r2)
```

## [문제 3] 같은 가족 조건에 부합하는 데이터 추출하기 2
### 3-1. df2의 인원 집계한 자료로 다음 조건을 처리하여 df3에 대입하라
1. 가족 조건이려면 2인 이상이어야 한다. 인원수(cnt)가 2명 이상만 추출한다. 
- 2명 이상만 추출한 결과 가족 조건에 부합하는 99개의 가족이 존재한다
2. 'Last'와 'Ticket'별 오름차순으로 정렬한다. 
3. 행의 index는 0부터 일련번호로 구성한다. 
- 다음은 df3의 결과 일부이다

### 3-2. df3['cnt']컬럼에서 10개만 추출한 Series를 r3에 대입하라
- 다음은 r3의 인쇄 결과이다


```python
# [문제 3]
#--[아래에 코드 작성]--------------------------------------------------
df3 = df2.query('cnt>=2').sort_values(['Last','Ticket']).reset_index(drop=True)
# df3 = df2[df2['cnt']>=2].sort_values(['Last','Ticket']).reset_index(drop=True)
r3 = df3['cnt'].head(10)

#--[검증용코드, 수정금지]--------------------------------------------------
df3
print(r3)
```

## [문제 5] 위에서 완성한 r4를 이용하여 같은 가족간 비교 분석
- 다음은 r4의 결과이다


### 1. 다음 조건에 따른 그룹을 r4에서 제거하라
- r4를 종합적으로 검토한 결과 Last, Ticket, Pclass가 모두 같고, SibSp와 Parch가 모두 0이 아니면 한 가족으로 정의할 수 있다. (SipSp는 형제자매수, Parch는 부모자식수를 의미)
- 13, 14행은 Last, Ticket, Pclass이 모두 같지만 SibSp와 Parch가 모두 0이므로 가족이 아니다. 
- Last, Ticket, Pclass 별 그룹 중에서 SibSp와 Parch가 모두 0인 사람들의 그룹을 제거하라.
- 단 삭제시 조건에 맞춰 해당 index를 찾아서 제거하라
- 결국 r4의 결과에서 Johnson가족 5명중 LINE 티켓에 탑승자 2명인 13, 14번 행이 삭제된다

### 2. 제거된 r4 데이터프레임으로 다음과 같이 집계하라
- 'Last'와 'Ticket'별 생존자수와 평균나이로 집계하여 r5에 대입하라 
- r5의 결과로 다음 그림을 참고하라


```python
# [5] 

#--[아래에 코드 작성]--------------------------------------------------

idx = r4.groupby(['Last','Ticket','Pclass']).filter(lambda x: all(x['Parch']+x['SibSp'])==0).index
r4.drop(index=idx, inplace=True)
r5 = r4.groupby(['Last','Ticket']).agg({'Age':'mean', 'Survived':'sum'})
#r5 = r4.pivot_table(index=['Last','Ticket'], aggfunc={'Survived':'sum', 'Age':'mean'})

#--[검증용코드, 수정금지]--------------------------------------------------
idx
print(r5)
```

## [문제 1] Age, Cabin 컬럼의 결측치를 다음 조건에 따라 처리하라
- df.isna().sum()를 실행하면 Age는 177개, Cabin은 687개 누락됨을 알수 있다
1. 'Cabin'컬럼은 호실번호이다. 호실번호의 첫글자로 수정하고 누락값(NaN)은 모두 'N'값으로 대체하라
    - 예로 'C85', 'B123', 'E46', NaN, ... 이면 'C', 'B', 'E', 'N', ... 로 수정한다
2. 'Age'컬럼은 결측치를 선실(Pclass)별 성별(Sex) 평균(mean)나이로 대체하라 
3. 완성한 df에서 Cabin, Age 컬럼만 10개 행을 추출한 데이터프레임을 r1에 대입하라
- 다음 그림은 r1을 인쇄한 결과이다


```python
# [1] 
df.isna().sum() # 컬럼별 누락개수 확인

#--[아래에 코드 작성]--------------------------------------------------

df['Cabin'] = df['Cabin'].fillna('N').str[0]
# df['Cabin'] = df['Cabin'].fillna('N').apply(lambda x: x[0])
df['Age'] = df.groupby(['Pclass','Sex'])['Age'].transform(lambda x: x.fillna(x.mean()))
r1 = df[['Cabin','Age']].head(10)

#--[검증용코드, 수정금지]--------------------------------------------------
df.head(6)
print(r1)
```
