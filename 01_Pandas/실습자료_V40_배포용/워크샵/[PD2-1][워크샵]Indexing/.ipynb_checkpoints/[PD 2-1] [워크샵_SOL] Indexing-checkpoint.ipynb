### [7] line차트로 시각화하기
1. plot() 메서드로 시각화 한다
    1. plot()의 이해  
        - 기본적으로 matplotlib 모듈을 이용하여 시각화한다  
        - sr.plot() ==> sr의 index가 x축, values가 y축에 line챠트로 시각화한다
    2. plot()의 옵션
        - figsize=(너비, 높이) : 챠트 크기를 지정하는 옵션으로 인치 단위  
        - xticks=sequence : x축상의 위치 표시 지점에 label을 직접 지정    
        - tick이란? 축상의 위치 표시 지점을 틱(tick)이라고 하고 이 틱에 써진 숫자 혹은 글자를 틱 레이블(tick label)이라고 한다. 틱의 위치나 틱 레이블은 자동으로 정해진다   
        
         
2. df2에서 인계동의 전연령대 Series를 line 차트로 시각화하는 다음 코드를 이해하라

```python
# line 챠트로 시각화
df2 = pd.read_csv('suwon_census.csv', index_col='동')
sr = df2.loc['인계동', '0세':]
sr

sr.plot(figsize=(15,5))
```

```python
# x축의 xtick을 10살 단위로 조정
sr.plot(figsize=(15, 5), xticks=range(0,101,10))
```

### [8] 특정 지역명을 입력받아 전연령층을  line 챠트로 시각화 하기
1. 지역명(동)을 dong 변수에 입력 받는다
2. df2에서 입력받은 동의 모든 연령을 인덱싱하여 sr 변수에 대입하라
3. sr로 line 챠트로 표시하라. 단, xtick을 10단위로 조정
   - 다음 그림은 지역명을 입력받아 시각화한 결과이다
  


```python
df2 = pd.read_csv('suwon_census.csv', index_col='동')
print('* 지역(동) 목록==>', df2.index.to_list())
dong = input('동명을 입력하시오==>')

# [8] 코드 작성
sr = df2.loc[dong, '0세':]
sr
sr.plot(figsize=(15, 5), xticks=range(0,101,10))
```
