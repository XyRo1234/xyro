### [1] 'stock.csv'와 'my_stock.csv' 파일 확인하기
- 참고 사이트 : http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201 
1. 코스닥 주식관련 데이터셋 'stock.csv' 파일(12행*8열)을 읽어 온 데이터프레임 df1이다
- 종목코드는 식별코드로 6자리로 이루어져 있고 문자를 포함하고 있어 문자열타입이다   
- 종목크드는 끝자리가 보통주 일경우 ‘0’, 우선주인 경우는 ‘5’, 'K', 'M'등으로 구성되어 있다
2. 관심있는 종목코드와 종목명으로 구성된 'my_stock.csv'파일(6행*2열)을 읽어온 데이터프레임 df2이다
- df1내에 존재하는 종목코드중 일부의 데이터로 구성되어 있다

```python
df1 = pd.read_csv('stock.csv', index_col='종목코드')
df2 = pd.read_csv('my_stock.csv', index_col='종목코드')
df1; df2
```

### [2] df1과 df2를 결합하여 mdf변수에 대입하라
- df2의 종목코드와 일치하는 행만 df1과 병합하라
- 다음 그림을 참고하여 같은 결과가 나오도록 구현하라


```python
# [2] 코드 작성
mdf = df2.join(df1.iloc[:, 1:], how='left')
# mdf = df2.merge(df1.iloc[:, 1:], left_index=True, right_index=True, how='left')
# mdf = df2.merge(df1, how = 'inner', on = ['종목코드','종목명'])
mdf
```

### [3] 위에서 완성한 mdf에서 다음 조건에 따라 그룹 집계하라
- '업종명'별 전 수치 컬럼에 대해 mean 집계 한다
- groupby()로 집계낸 결과를 x에 대입하여 확인하라
- pivot_table()로 집계낸 결과를 y에 대입하여 확인하고 x와 차이를 확인하라
- x, y의 dtypes로 확인하고 groupby와 pivot_table의 동작 차이를 검토하여 보자


```python
# [3] 코드 작성
x = mdf.groupby('업종명').mean()
# x = mdf.groupby('업종명').mean().astype({'거래대금':np.int64})
y = mdf.pivot_table(index='업종명')
x; x.dtypes
y; y.dtypes
```

### [4] 위에서 완성한 mdf에서 다음 조건에 따른 그룹집계 결과를 r4에 대입하라
- 아래 주어진 dic은 종목코드를 키로 그룹명을 value로 구성되어 있다
- 주어진 dic으로 구성된 2개의 그룹별('G1', 'G2') max 집계를 하려 한다
- '종가' '등락률', '시가' 컬럼에 대해 max 집계를 적용하여 r4에 대입하라
- 코드 실행 결과는 다음 그림과 같아야 한다


```python
dic = {'282330':'G1', '128820':'G1', '145995':'G2','336370':'G2'}

# [4] 코드 작성
r4 = mdf.groupby(dic)[['종가', '등락률', '시가']].max()
r4
```

### [5] 위에서 완성한 mdf에서 다음 조건에 따른 그룹집계 결과를 r5에 대입하라
- '등락률' 컬럼에서 등락률이 +인 경우와 -인 경우로 그룹화하여 그룹별 개수로 집계하라
- 단, 등락률이 +이면 'up', -이면 'down' 레벨명으로 지정한다
- (참고) groupby 객체에 size() 메서드를 적용하면 그룹별 행의 수를 집계한다


```python
# [5] 코드 작성
arr = np.where(mdf['등락률']>0, 'up', 'down')
arr
r5 = mdf.groupby(arr).size()
r5
```

### [6] 위에서 완성한 mdf에서 다음 조건에 따른 그룹집계 결과를 r6에 대입하라
- '종목코드'에서 보통주와 우선주로 그룹화하여 그룹별 개수로 집계하라
- (참고) 종목코드는 끝자리가 보통주 일경우 ‘0’, 우선주인 경우는 ‘5’, 'K', 'M'등으로 구성되어 있다


```python
# [6] 코드 작성
r6 = mdf.groupby(lambda x: '보통주' if x.endswith('0') else '우선주').size()
r6
```
