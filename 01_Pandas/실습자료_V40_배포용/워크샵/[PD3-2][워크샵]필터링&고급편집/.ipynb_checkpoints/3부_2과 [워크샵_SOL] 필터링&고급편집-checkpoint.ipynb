### [5] 총인구수로 3등급으로 분류하기
#### 다음에 주어진 조건에 따라 'level '컬럼을 '총인구수'앞 컬럼에 추가하라
- 총인구수로 4만명이상이면 'high', 2만명~4만명미만이면 'mid', 2만명미만이면 'low'로 평가한다



```python
df = pd.read_csv('suwon_census4.csv')

# [5] 코드 작성 
def func(x):
    if x>=40000 : return 'high'
    elif x>=20000: return 'mid'
    else: return 'low'
    
df.insert(3, 'level', df['총인구수'].map(func))
# df.insert(3, 'level', df['총인구수'].apply(func))
df.head(3)
```

### [6] 연령단계를 3개 그룹의 집단으로 재구성하라
#### 다음 주어진 조건에 따라 새로운 데이터프레임 df2를 생성하고 인쇄하라
1. 아동, 성인, 노인 3개 그룹의 연령별 합산 결과로 새로운 데이터프레임을 df2에 대입한다
- 주어진 func함수를 이용하여 df2를 생성한다
    - func()함수는 아동(0-19), 성인(20-59), 노인(60-) 3단계로 구분하여 합산한 목록을 리턴한다
2. df2의 집계된 3개의 컬럼명을 ['아동','성인','노인']으로 지정한다
3. '동' 컬럼을 첫번째 컬럼에 위치하도록 하며 그 뒤에 집계한 3개 컬럼을 위치 시킨다
- 다음 그림을 참고하고 일부만 인쇄한 결과이다


```python
df = pd.read_csv('suwon_census4.csv')

def func(x):
    return [ x['0~9세':'10~19세'].sum(), x['20~29세':'50~59세'].sum(),x['60~69세':].sum() ]

# [6] 코드 작성
df2 = pd.DataFrame(df['동'])
df2[['아동', '성인', '노인']] = df.apply(func, axis=1, result_type='expand')
df2.head()
```

### [7] 동별 총인구수를 수원시 총인구수 대비 백분율로 계산하라
#### 지역별 수원시 인구 대비 백분율을 구한 '인구비율'컬럼을 '총인구수'컬럼 뒤에 추가하라
- '인구비율' = 동별 총인구수 / 수원시 총인구수
- 다음 그림의 출력 결과를 참조하라


```python
df = pd.read_csv('suwon_census4.csv')

# [7] 코드 작성
df.insert(4, '인구비율', df['총인구수'].apply(lambda x: '{:.2%}'.format(x/df['총인구수'].sum())))
df.head(3)
```

### [8] '총인구수'가  3만명 미만일 경우 'low'로 처리하라
#### 총인구수가 3만명 미만일 경우 'low'로 처리한 새로운 '총인구수2' 컬럼을 '총인구수' 컬럼 뒤에 추가하라
- 단, 3만명 이상일 경우 원래 인구수 값을 유지한다. 다음 그림을 참조하라


```python
df = pd.read_csv('suwon_census4.csv')

# [8] 코드 작성
df.insert(4, '총인구수2', df['총인구수'].mask(df['총인구수'] < 30000, 'low'))
df.head(3)
```
