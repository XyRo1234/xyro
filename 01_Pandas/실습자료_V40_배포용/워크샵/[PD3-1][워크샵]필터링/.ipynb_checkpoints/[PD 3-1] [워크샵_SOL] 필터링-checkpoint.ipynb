### 실습에 사용할 DataFrame
- 아래 df는 원본 DataFrame인 tdf에서 6개의 컬럼만 읽어온 DataFrame이다
- 앞으로 실습은 이 df를 사용하여 진행한다
- 이 df의 구성을 확인하라

```python
df = pd.read_csv('titanic_train.csv', usecols=[1,2,3,4,5,7]) 
df
```

### [2] 60세 이상 생존자의 명단을 추출하라
- 생존자의 경우는 Survived 컬럼의 값이 1이다
- 추출한 DataFrame은 변수 r2에 대입하고 내용을 확인하라
- 아래 그림은 추출한 결과이다


```python
# [2] 코드 작성
r2 = df[ (df['Survived']==1) & (df['Age']>=60) ]
r2
```

### [3] 생존자중 나이가 많은 상위 3명의 행을 추출하라
- 추출한 명단은 변수 r3에 대입하고 내용을 확인하라
- 아래 그림은 추출한 DataFrame이다


```python
# [3] 코드작성
r3 = df[ df['Survived']==1 ].nlargest(3, 'Age')
r3
```

### [4] 'Name' 컬럼에서 호칭이 'Miss'인 여성중 나이가 20세 이상인 생존자의 인원수를 인쇄하라
- 결과는 60이 인쇄되어야 한다

```python
# [4] 코드작성
r4 = df.loc[ df['Name'].str.contains('Miss')&(df['Age']>=20)&(df['Survived']==1),'Name'].count()
# r4=df.loc[ df['Name'].str.contains('Miss')&(df['Age']>=20),'Survived'].sum()
r4
```

### [5] 성별 생존율을 pie챠트로 시각화하라
#### 성별 생존자의 인원수를 집계낸 Series를 이용하여 pie챠트로 시각화하라
- 다음 집계결과와 챠트 실행 결과를 참고하라


```python
# [5] 코드작성
r5 = df.loc[ df['Survived']==1, 'Sex'].value_counts()
r5

# pie챠트 시각화
plt.title('성별 생존율')
r5.plot(kind='pie', figsize=(7,7), autopct='%1.1f%%', explode=[0.1,0], shadow=True)
```

### [6] 생존 여부(생존자수와 사망자수)를 pie 챠트로 시각화하라
#### 아래 요구한 코드를 설계한 후 pie 챠트로 시각화하라 
1. 생존 여부에 따라 'alive', 'dead'값을 갖는 새로운 컬럼 'Survived2'를 추가하라
2. 'Survived2'컬럼으로 생존 여부를 이용하여 pie 챠트로 시각화하라
- 아래 그림은 df의 일부와 pie챠트이다. 다음 그림을 참고하라


```python
df = pd.read_csv('titanic_train.csv', usecols=[1,2,3,4,5,7])

# [6] 코드 작성
df['Survived2'] = df['Survived'].replace({0:'dead', 1:'alive'})
df
r6 = df['Survived2'].value_counts()
plt.title('생존여부별 생존율')
r6.plot(kind='pie',  figsize=(6,6), autopct='%.1f%%', explode=[0.1, 0])
```

### [7] 나이 정보가 누락된 사람들의 선실 및 성별 분포도를 확인하려 한다
1. df의 'Age' 컬럼 값이 누락된 행들을 추출하여 r7에 대입하라
2. r7에서 선실별 및 성별 인원수(빈도 집계)을 구하여 r8에 대입하라
3. r8을 이용하여 다음 그림과 같이 pie챠트로 시각화하여 분포도를 확인하라
- 다음 그림을 참고하라



```python
r7 = df[df['Age'].isna()]
r8= r7[['Pclass','Sex']].value_counts()
r8

r8.plot(kind='pie', figsize=(7,7), autopct='%1.1f%%',\
        explode=[0.1,0,0,0,0,0], shadow=True, title='선실별 및 성별 인원비율')
```
